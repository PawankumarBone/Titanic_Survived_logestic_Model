```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
tt=pd.read_csv('train.csv')
tt
```

```python
tt[tt['Fare']>250].groupby(['Sex','Embarked'])['PassengerId'].count().plot(kind='bar',color=['orange','green','black','pink'])
```

```python
tt
```

```python
tt.isnull().sum()
```

```python
tt=tt.fillna(tt.mean())
```

```python
tt
```

```python
tt.drop(['Cabin'],axis=1)
```

```python
tt=tt.drop(['Cabin','Name'],axis=1)
tt
```

```python
col=['Age','Fare']
for c in col:
  percentile25=tt[c].quantile(0.25)
  percentile75=tt[c].quantile(0.75)
  iqr=percentile75-percentile25
  upper_limit=percentile75+(1.5*iqr)
  lower_limit=percentile25-(1.5*iqr)
  tt=tt[(tt[c]<upper_limit) & (tt[c]>lower_limit)]
  plt.figure()
  sb.boxplot(y=c, data=tt)
```

```python
tt
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
col=['Sex','Ticket','Embarked']
```

```python
for c in col:
    tt[c]=le.fit_transform(tt[c])
tt
```

```python
x=tt.drop(['Survived'],axis=1)
x
```

```python
y=tt['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=30)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg=LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
y_pred=reg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
tn,fp,fn,tp=confusion_matrix(y_test,y_pred).ravel()
tn,fp,fn,tp
```

```python
sb.heatmap(confusion_matrix(y_test,y_pred),annot=True)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
